```python
import yfinance as yf

# Download S&P 500 index data
sp500 = yf.download('^GSPC', start='2014-01-01', end='2024-01-01')

# Preview the data
print(sp500.head())
```

```python
import pandas as pd

sp500['Daily Return'] = sp500['Close'].pct_change()
sp500['Cumulative Return'] = (1 + sp500['Daily Return']).cumprod()
sp500['20D MA'] = sp500['Close'].rolling(window=20).mean()
```

```python
mean_return = sp500['Daily Return'].mean()
volatility = sp500['Daily Return'].std()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sp500['Close'].plot(figsize=(10, 6), title='S&P 500 Close Price')
plt.show()

sns.histplot(sp500['Daily Return'], bins=50, kde=True)
plt.title("Distribution of Daily Returns")
plt.show()
```
